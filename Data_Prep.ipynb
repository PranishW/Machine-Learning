```python
X = df.iloc[:, 1:6]
y = df["AHD"]
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.75, test_size=0.25, random_state=42)
print("X train : ");
print(X_train);
print("\nX test : ");
print(X_test);
print("\nY train : ");
print(y_train);
print("\nY test : ");
print(y_test);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
plt.bar(df['ChestPain'] , df["Ca"])
plt.legend(loc='upper left')
plt.grid(True, linewidth= 1, linestyle="--")
plt.xlabel("Chest Pain Types")
plt.title("Rest BP/Max HR vs Age")
```

```python
plt.figure(figsize=(15, 10))
plt.bar([a-0.5 for a in df["Age"]], df["RestBP"], width= 0.5, label = 'Rest BP', align='edge')
plt.bar([a+0.5 for a in df["Age"]], df["MaxHR"], width= -0.5, label = 'Max HR', align='edge')
plt.legend(loc='upper left')
plt.grid(True, linewidth= 1, linestyle="--")
plt.xlabel("Age")
plt.title("Rest BP/Max HR vs Age")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=df["Age"], y=df["Chol"]).set(title='AHD Presence of Cholestrol vs Age')
```
